```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 4, 8, 16, 32, 64, 128])
N20_M = np.array([0.000077, 0.000131, 0.000188, 0.000321, 0.000568, 0.001615, 0.002678, 0.014115])
N20_sd = np.array([0.00002162817093, 0.00001663329993, 0.00001813529401, 0.000030713732,
                   0.00005996295152, 0.001883821471, 0.00195311034, 0.0006264937705])

N100_M = np.array([0.004304, 0.002568, 0.001711, 0.001231, 0.001108, 0.001282, 0.005602, 0.014515])
N100_sd = np.array([0.000287100447, 0.0002284634666, 0.0002009118104, 0.0001132793989,
                    0.00009681597664, 0.00006908931418, 0.007344017671, 0.0019372274])

N1000_M = np.array([3.705663, 1.85473, 0.927986, 0.46453, 0.235451, 0.123978, 0.074728, 0.070175])
N1000_sd = np.array([0.01348438195, 0.00868902686, 0.003950519516, 0.00197663575,
                     0.001546713002, 0.002255860516, 0.01095301562, 0.0035893585])
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 5))
ax[0].errorbar(X, N20_M, yerr=N20_sd, label='strong scaling', marker='.', ls='--')
ax[0].plot(X, N20_M[0]/X, label='ideal scaling', marker='.', alpha=0.8)
ax[0].set_xscale("log", base=2)
ax[0].set_yscale("log", base=10)
ax[0].set_xlabel("Number of OMP Threads")
ax[0].set_ylabel("Total Runtime (s)")
ax[0].set_title("N = 20")
ax[0].legend()

ax[1].errorbar(X, N100_M, yerr=N100_sd, label='strong scaling', marker='.', ls='--')
ax[1].plot(X, N100_M[0]/X, label='ideal scaling', marker='.', alpha=0.8)
ax[1].set_xscale("log", base=2)
ax[1].set_yscale("log", base=10)
ax[1].set_xlabel("Number of OMP Threads")
ax[1].set_ylabel("Total Runtime (s)")
ax[1].set_title("N = 100")
ax[1].legend()

ax[2].errorbar(X, N1000_M, yerr=N1000_sd, label='strong scaling', marker='.', ls='--')
ax[2].plot(X, N1000_M[0]/X, label='ideal scaling', marker='.', alpha=0.8)
ax[2].set_xscale("log", base=2)
ax[2].set_yscale("log", base=10)
ax[2].set_xlabel("Number of OMP Threads")
ax[2].set_ylabel("Total Runtime (s)")
ax[2].set_title("N = 1000")
ax[2].legend()
```
